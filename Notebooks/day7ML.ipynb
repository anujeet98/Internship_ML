```python
import pandas as pd#for data frame
from mlxtend.preprocessing import TransactionEncoder  #used for converting user data like 'Cola' into number or any form that computer understands
from mlxtend.frequent_patterns import apriori  #apriori is the algorithm that we will use here

dataset=[["Bread","Milk","Beer"],
         ["Bread","Diapers","Eggs"],
         ["Milk","Diapers","Beer","Cola"],
         ["Bread","Milk","Diapers","Beer"],
         ["Bread","Milk","Cola"]]
```

```python
te=TransactionEncoder()#create instance of engine
te=te.fit(dataset)#Encode
te_ary=te.transform(dataset)#transforms into columns
print(te.columns_)
print(te_ary)
```

```python
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
#finds frequency of the items
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets,metric="support",min_threshold=0.4)
rules
```

```python
rules[rules['support']==rules['support'].max()]
```
