```python
#make a year column using function and using one line lambda
#to add year col
def getyear(x):
    return int(x.split("/")[-1])
data['year']=data['Date'].apply(lambda x:int(x.split("/")[-1]))
data.head()
```

```python
#make separate column for date and month
data['year']=data['Date'].apply(lambda x:int(x.split("/")[-1]))
data['month']=data['Date'].apply(lambda x:int(x.split("/")[0]))
data['date']=data['Date'].apply(lambda x:int(x.split("/")[1]))
data.head()
```

```python
data['year_new']=data['Date'].apply(getyear)
data.head()
```
