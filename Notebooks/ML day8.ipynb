# Linear Regression

# predict pizza price based on pizza size

```python
%matplotlib inline
import matplotlib.pyplot as plt

#list of list input and output format is important
#Ordinary Least Square takes multiple inputs and predicts multiple outputs

Psize=[[6],[8],[10],[14],[18]]#input hasa list of list
Pprice=[7,9,13,17.5,18]

plt.figure()
plt.title('pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(Psize,Pprice,'go')#go=color green,marker style o
plt.axis([0,25,0,25])#xlims and ylims together
plt.grid(True)
#here we got 5 points from the data
```

```python
from sklearn.model_selection import train_test_split  #to split data into test and train data
Psize_train,Psize_test,Pprice_train,Pprice_test=\
train_test_split(Psize,Pprice,test_size=0.2,random_state=20)#taking variables Psize_train,Psize_test,Pprice_train,Pprice_test
#separating 20% as test case ie 1 out of 5 and 80% as train case ie 4out of 5
#if we use random_state=20 then every time re running won't give different test point
plt.figure()
plt.title('pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(Psize_train,Pprice_train,'r*')#go=color red,marker style *
plt.axis([0,25,0,25])#xlims and ylims together
plt.grid(True)
#here plotting 4 points only train points
```

```python
from sklearn.linear_model import LinearRegression  #to use linear Regression
model=LinearRegression()  #model is m and c in y=mx+c
model.fit(Psize_train,Pprice_train)#fitting train data

print(Pprice_test)
model.predict(Psize_test)
```

```python
#plot initial data
plt.figure()
plt.title('pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(Psize_train,Pprice_train,'go')
plt.plot(Psize_test,Pprice_test,'bo')# go for plotting train case and blue (o)for plotting test case

plt.axis([0,25,0,25])#xlims and ylims together
plt.grid()

#plot regression solution line
regression_line_y=model.predict([[0],[25]])
print(regression_line_y)
plt.plot([0,25],regression_line_y,'r',linestyle='-',linewidth=1)
```

```python
model.predict(15)
```

```python
model.score(Psize,Pprice)
```

```python
print("slope(m):%2f,intercept(c):%2f"%(model.coef_,model.intercept_))
```
