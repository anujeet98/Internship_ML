# boolean indexing

```python
score_filter=reviews["score"]>7
score_filter
```

```python
myDf=reviews.head()
myDf[[True,False,False,True,True]]#prints all the rows where it is true
```

```python
reviews[score_filter].head()# returns dataframe
#gets head (top5) dataset of filter
```

```python
reviews[score_filter]#gives all the filter columns
```

```python
reviews[reviews["platform"]=="PC"].shape 
```

```python
#using & operator for multiple filter option
xbox_one_filter=(reviews["score"]>7) & (reviews["platform"]=="Xbox One")
print(reviews[xbox_one_filter].shape)
reviews[xbox_one_filter]
```

```python
#use proper variable name to make it easy to understand
is_high_score=(reviews["score"]>7)
Is_xbox_one=(reviews["platform"]=="Xbox One")
reviews[is_high_score & Is_xbox_one]
```

# grouping data

```python
grouped_data=reviews.groupby(["platform"])

for name,group in grouped_data:
    print("name:%s, Count:%s"%(name,group['score'].count()))
print("_______________________")
print(grouped_data.get_group('Android'))
```
